```python
from ultralytics import YOLO

print("Loading base model...")
model = YOLO("yolo11s.pt")  # Downloads automatically

print("Starting training...")
try:
    # Training parameters (adjust as needed)
    results = model.train(
        data="dataset/data.yaml",
        epochs=50,           # Increase to 100+ for better accuracy
        imgsz=640,          # Image size (must be multiple of 32)
        patience=10,        # Early stopping patience
        device=0,           # GPU device (0 = first GPU)
        batch=16,           # Batch size (adjust based on GPU memory)
        amp=True,           # Automatic Mixed Precision for faster training
    )
    
    print("\n✓ Training complete!")
    print(f"Best model saved at: {results.save_dir}/weights/best.pt")
    
except Exception as e:
    print(f"✗ Training failed: {e}")
    print("Make sure your 'data.yaml' path is correct and dataset is properly formatted.")
```
